# 07 · Stats & results

Headline numbers over the unified `output/` root: produced books, pages, the
**distribution** size (access copies) and the **archival** size, throughput and
ETA. Reads every `batch_report*.jsonl`, keeping the newest row per book.

```python
import sys
from pathlib import Path

# project importable when the kernel isn't the project .venv
sys.path.insert(0, str(Path.cwd().parent))

from evilflowers_books_digitalizer.runtime import load_runtime
from evilflowers_books_digitalizer.cache import LocalCache

# ONE source of truth for cache_dir + output_dir (paths resolve to the project
# root regardless of the notebook's working directory) — this is the unified
# cache every notebook and the production runner share.
rt = load_runtime()
cache = LocalCache(rt.cache_dir)
print("cache :", rt.cache_dir)
print("output:", rt.output_dir)
```

```python
from evilflowers_books_digitalizer.reporting import (
    load_reports, latest_per_book, summarize_reports, summarize_by_source)

rows = latest_per_book(load_reports(rt.output_dir))
overall = summarize_reports(rows)
print("books:", overall["books"], "| produced:", overall["produced"])
print("pages:", overall["pages"], "| distribution:", overall["pdf_mb"], "MB",
      "| archival:", overall.get("archival_mb"), "MB")
print("MB/page:", overall["mb_per_page"], "| pages/min:", overall["pages_per_min"])
overall["by_status"]
```

```python
import pandas as pd
by_src = summarize_by_source(rows)
pd.DataFrame(by_src).T[["books", "produced", "pages", "pdf_mb", "archival_mb", "minutes"]]
```
